### c) Bilateral filtering

```python
def weight_function(img, wsize, sigma_d, sigma_r, i, j, k, l):
    """ Calculate weights for the bilateral filter
    Parameters:
        i,j: indices
        k,l: indices
    Returns:
        weights: weights for bilateral filter
    """
    # domain kernel, formula (3.35)
    domain_kernel = np.exp((-1) * ((i-k)**2 + (j-l)**2) / (2 * sigma_d**2))
    
    # range kernel, formula (3.36)
    range_kernel = np.exp((-1) * np.sqrt((img[i][j]**2 - img[k][l]**2)**2) / (2 * sigma_r**2))
    
    # calculate weights
    weights = domain_kernel * range_kernel
    
    return weights

def bilateral_filter(img, wsize, sigma_d, sigma_r):
    """ Calculate bilaterial filter, formular (3.34)
    Parameters:
        img: noisy image
        wsize: weight size
        sigma_d: smoothing parameter
        sigma_r: smoothing parameter
    
    Returns:
        output: img with the denoised intensity/value of the pixel
    """
    img = img.copy()
    radius = int(0.5 * wsize) # get the half of the weight size
    image_length_x = img.shape[0]
    image_length_y = img.shape[1]
    
    for i in range(radius, image_length_x - radius): # we don't start at index 0, otherwise the filter goes out of bounce. We start at half length of the weights. 
        for j in range(radius, image_length_y - radius): # same applies to the vertical axes
            weights_sum = 0 # denominator of the fraction, formula (3.38)
            weighted_intensity = 0 # nominator of the fraction (3.38)
            
            for iteration in range(-1*radius, radius+1): # we iterate through all cells of the weight grid
                k = i+iteration # get the indices for each iteration
                l = j+iteration
                weights = weight_function(img, wsize, sigma_d, sigma_r, i, j, k, l) # calculate the weights
                weights_sum += weights # sum up the weights
                weighted_intensity += img[k][l] * weights # calculate the weighted intensity
                
            new_pixel = weighted_intensity / weights_sum
            img[i][j] = new_pixel
    
    return img
```
